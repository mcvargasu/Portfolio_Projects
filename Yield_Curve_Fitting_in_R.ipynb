# **Fitting the yield curve using cubic splines**


```r
#Treasury Bond T 2 3/8 11/15/49
```

```r
t1 = c(0.5,1,2,4,6,10,14,20,40,60)
y1 = c(0.0014,0.0019, 0.0019, 0.00245, 0.0029 ,0.0035 ,0.0045,0.0047,0.00655,0.0078)
```

```r
spl1 <- smooth.spline(y1 ~ t1)
```

```r
plot(spl1, ylab = 'Yield Treasury 1', xlab = 'Years', main = 'Treasury Yield Curve for 13/03/2020')
lines(spl1)
```

```r
predict(spl1, t1)
t1.new <- seq(from = 0.3461538, to = 59.3461538, by = 1)
predict(spl1, t1.new)->NewGraph1
plot(NewGraph1, ylab = 'Yield', xlab = 'Years', main = 'Treasury Yield Interpolated Curve for 13/03/2020')
lines(NewGraph1)
```
